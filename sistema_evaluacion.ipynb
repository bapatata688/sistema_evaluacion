Cada rama corresponde a una seccion del proyecto, lo que permite trabajar de forma paralela e independiente por area.


````python
import os

# Recorre y muestra la estructura de archivos del proyecto
def mostrar_estructura(ruta_base="."):
    for raiz, carpetas, archivos in os.walk(ruta_base):
        nivel = raiz.replace(ruta_base, "").count(os.sep)
        sangria = "    " * nivel
        print(f"{sangria}{os.path.basename(raiz)}/")
        for archivo in archivos:
            print(f"{sangria}    {archivo}")

mostrar_estructura(".")
````
